### This notebook does some analysis uisng the saved model and datasests

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
import pickle
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

### Load model and make predictions

```python
# read test/submission dataframe
df_test = pd.read_pickle("../data/df_test.pkl")
```

```python
# read saved model
model = joblib.load("../model/model_latest.pkl")
```

```python
label_col = ["item_sales"]
trainCols = df_test.columns[~df_test.columns.isin(label_col)]
X_test = df_test[trainCols]
y_test = df_test[label_col]
```

### Feature importance

```python
feature_imp = pd.DataFrame(sorted(zip(model.feature_importances_,X_test.columns)), columns=['Value','Feature'])
plt.figure(figsize=(20, 10))
sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False))
plt.title('LightGBM Features (avg over folds)')
plt.tight_layout()
plt.show()
plt.savefig('lgbm_importances-01.png')
```

### Compare rmse between lightgbm and naive model


For the sake of this exericse I am using rmse for error/performance. With more time a better metric ccould have been chosen.

```python
def rmse(predictions, targets):
    """root mean square error"""
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
def naive_model(df):
    """
    Naive model using the last actual value copied
    for all items
    Input: Dataframe: combined sales dataframe
    Returns: actual and forecast values by date and item
    """
    # Get last date for sales dataframe
    sales_df_last = sales_df[sales_df['date'] == '2016-04-23']
    sales_df_last = sales_df_last[['id', 'date', 'item_sales']]
    # Get values from test dtataframe
    sales_df_test = sales_df[(sales_df['date'] >= '2016-04-04') \
                             & (sales_df['date'] <= '2016-05-22')]
    sales_df_test = sales_df_test[['id', 'date', 'item_sales']]
    final_df = pd.merge(sales_df_test, sales_df_last, how='left', on=['id'])
    final_df = final_df.rename(columns={'item_sales_x' : 'actuals', \
                                       'item_sales_y' : 'predictions', 'date_x' : 'date'})
    final_df = final_df.drop(['date_y'], axis=1)
    return final_df
    
    
```

```python
sales_df = pd.read_pickle("../data/sales_df.pkl")
final_df = naive_model(sales_df)
```

```python
# make prediction using previuosly trained model
# and print rmse
y_pred = model.predict(X_test)
y_test_ar = np.array(y_test["item_sales"])
print("Lightgbm model rmse is:", rmse(y_pred, y_test_ar))
```

```python
# rmse for naive model
y_test = np.array(final_df['actuals'])
y_pred = np.array(final_df['predictions'])
print("Naive model rmse is:", rmse(y_pred, y_test))
```

So lightgbm model performs beteer for this dataset
